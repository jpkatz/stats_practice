```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create a PMF
# assuming a company has a package failure rate following a beta distribution
```

```python
alpha = 1.0
b = 0.5
x = np.linspace(0, 1, 101)
data = beta.pdf(x, alpha, b)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, beta.pdf(x, alpha, b),'r-', lw=5, alpha=0.6, label='beta pdf')
```
